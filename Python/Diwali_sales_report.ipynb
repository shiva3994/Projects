### Missing value and Anamolous detection

```python
df.isnull()
# check for null
```

```python
df.isnull().sum()
# check for sum of null columnwise
```

```python
null_counts = df.isnull().sum()
null_counts
```

```python
null_counts = null_counts[null_counts > 0]
null_counts
# show column containing nullhaving value above 0
```

```python
plt.figure(figsize=(10,5))
null_counts.plot(kind='bar',color='green')

plt.title('Missing value per column') # title
plt.xlabel('Column name') # X axis name
plt.ylabel('Number of missing values') # Y axis name

plt.xticks(rotation = 40)
plt.yticks(rotation = 40)

plt.show() # show final graph 
```

```python
cat_cols = ['Status','unnamed1']
for col in cat_cols:
    df[col] = df[col].fillna('unknown')
# Resolve the categorical with unknown

num_cols = ['Amount']
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
# Resolve the numerical with median
```
