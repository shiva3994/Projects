```python
# Task 1.3 – Missing & Anomaly Detection

# Quantify nulls per column (count and percentage)
null_counts = df.isnull().sum()
# Count how many empty boxes are in each column.
# df.isnull() looks at every single cell in the dataframe and flags it as True if it's empty
# Adding .sum() counts all those True flags per column.

null_pct = (null_counts / len(df) * 100).round(2)
# Calculate what percentage of the total data is missing for each column.
# To turn a raw count into a percentage 
        # You divide the number of missing cells (null_counts) by the total number of rows in the entire spreadsheet (len(df)).
        # Multiplying by 100 converts that decimal into a standard percentage format (like 11.69 instead of 0.1169).
        # .round(2) chops off any massive trailing decimals so it looks clean, leaving just two numbers after the decimal point.

null_summary = pd.DataFrame({'null_count': null_counts, 'null_pct': null_pct})
# Glue the counts and the percentages together into a fresh, readable table.
# This line builds a brand new mini-spreadsheet (pd.DataFrame) called null_summary.
# It creates a column named 'null_count' and fills it with your raw counts, and a second column named 'null_pct' filled with your percentages.

null_summary = null_summary[null_summary['null_count'] > 0].sort_values('null_count', ascending=False)
# Filter out the perfect columns and sort the broken ones from worst to best.
# Filtering [...] # It tells Python, "I only care about columns where the missing count is greater than 0." 
                  # If a column is completely perfect (like year), it gets hidden so it doesn't clutter your view.

                  #       The Table                   The Rule Book
                    #          ↓                             ↓
                    #    null_summary [ null_summary['null_count'] > 0 ]

# Sorting .sort_values(...): # It sorts the remaining messy columns by their raw count.
                             # Setting ascending=False ensures the column with the absolute most missing data appears right at the very top.

                             #      sort_values('null_count', ascending=False)
                             # 'null_count': This tells pandas which column to use as the baseline for sorting.                             #    
                             # acending=False: By default, sorting goes from smallest to largest (1, 2, 3..). By setting ascending=False. It forces the largest numbers to the top.
                             
null_summary
# Print out the final table on my screen.
# In a Jupyter Notebook, simply writing the name of a variable as the last line of a code cell tells Python to display it nicely.
```

```python
# Anomaly detection: state codes should be 2-3 characters; anything longer is invalid (e.g. a VIN)

anomalous_state_mask = df['state'].astype(str).str.len() > 3

# This line creates a "mask"—which is essentially a checklist of True and False values matching every row in your dataset.
# df['state']: Targets the column named state.
# .astype(str): Forces everything in that column to be treated as text (a string), just in case a row accidentally contains a number.
# .str.len() > 3: Measures the length of the text in each row and asks, "Is this longer than 3 characters?"
                # If a row contains CA (length 2), it returns False (it's fine).
                # If it accidentally contains a 17-character Vehicle Identification Number (VIN), it returns True (it's an anomaly).

print(f"Anomalous 'state' values found: {anomalous_state_mask.sum()}")

# Counts up how many bad rows were found and prints the result.
# In Python, True equals 1 and False equals 0. By using .sum() on your mask, it adds up all the True rows. 
# The f"..." syntax lets you dynamically drop that final count straight into your printed sentence.


# Treat these anomalies as missing values
df.loc[anomalous_state_mask, 'state'] = np.nan

# df.loc[...]: This is a locator tool. It says, "Look at the dataset, filter it down to only the rows where my mask is True, and target the state column."
```

```python
# Find categorical columns with nulls and fill with 'Unknown'

categorical_fill_unknown = ['make', 'model', 'trim', 'body', 'color', 'interior', 'transmission']
for col in categorical_fill_unknown:
    df[col] = df[col].fillna('Unknown')
# col is temp name for column
# The line for col in categorical_fill_unknown: tells the computer: "Go through this list one by one. Each time you grab a name, temporarily call it col."
# df[col]: This looks at the specific column we are currently dealing with (for example, the color column).
# .fillna('Unknown'): This tells Python to scan all the way down that column, look for any blank/null spaces, and temporarily substitute them with the word 'Unknown'.
# Words or categories—like car brands (make), paint colors (color), or transmission types. 
# You can't calculate the mathematical average of "Red," "Blue," and "Silver."
# It loops through all these word-based columns and replaces any blank space with the word 'Unknown'.
# If 10% of the cars are missing a transmission type, throwing those rows away would waste a lot of good data.
# Labeling them "Unknown" keeps the rows intact while acknowledging that we just don't know that specific detail.

# Find Numerical columns with nulls and fill with median (robust to outliers/skew)
for col in ['condition', 'odometer', 'mmr']:
    df[col] = df[col].fillna(df[col].median())
# for col in [...]: It loops through three continuous numerical features: 
# vehicle physical condition rating (condition), mileage (odometer), and the Manheim Market Report wholesale value (mmr).
# df[col].median(): It calculates the middle value of the column when all numbers are sorted in order.
# .fillna(...): It scans the column for any blank/null cells and swaps them out for that calculated median.

# Drop the small number of rows where price / saledate / vin / state are missing
df = df.dropna(subset=['sellingprice', 'saledate', 'vin', 'state'])
# Pandas operations usually return a copy of the modified data.
# Re-assigning it back to df mutates your working DataFrame so that downstream analysis only runs on clean, fully-populated records.
# subset=[...]: By default, .dropna() will drop a row if any column in the entire DataFrame contains a null. 
# Passing a list to subset restricts this trigger exclusively to these four columns. If a row has a missing color but a valid price, VIN, date, and state, it will safely be kept.

print("Remaining nulls per column:")
print(df.isnull().sum().sum(), "total null values left")
# df.isnull(): Creates a matrix of Booleans (True for empty cells, False for populated cells) matching the exact shape of your DataFrame.
# .sum() (First sum): Vectorizes the operation vertically. It counts the True values column by column.
# Returning a pandas Series where the index is the column names and the values are the counts of missing data per column.
# .sum() (Second sum): Adds up all the column totals from that Series into one single integer representing every single missing cell in the entire dataset.

df.isnull().sum()
# df.isnull() (The Map)
# This scans your entire DataFrame cell by cell and converts it into a matrix of Boolean values (True or False).
# If a cell contains a missing value (NaN, None, or NaT), it becomes True.
# If a cell contains valid data, it becomes False.
```

```python
# Map known abbreviations/typos to their standard brand name, then title-case everything

make_map = { 'vw': 'Volkswagen', 'chev truck': 'Chevrolet', 'dodge tk': 'Dodge',
    'ford tk': 'Ford', 'ford truck': 'Ford', 'gmc truck': 'GMC',
    'hyundai tk': 'Hyundai', 'landrover': 'Land Rover', 'mazda tk': 'Mazda',
    'mercedes': 'Mercedes-Benz', 'mercedes-b': 'Mercedes-Benz'}
# This is a standard Python dictionary acting as a translation key. 
# The messy, abbreviated, or mistyped versions are on the left (keys), and the clean, official brand names you want to replace them with are on the right (values).

df['make'] = df['make'].str.lower().replace(make_map).str.title()

# .str.lower(): Converts every single car make in the column to lowercase (e.g., "VW", "Vw", and "vw" all become exactly "vw"). 
# This makes matching predictable because Python is case-sensitive.
# .replace(make_map): This is the heavy lifter. It looks at the lowercase text and checks if it matches anything in your make_map. 
# If it sees "ford tk", it swaps it out for "Ford". If it sees "mercedes-b", it swaps it for "Mercedes-Benz". If a name isn't in the map (like "Toyota"), it leaves it alone.
# .str.title(): Capitalizes the first letter of every word and lowercases the rest (e.g., "land rover" becomes "Land Rover"). 
# This ensures all your final data looks visually clean and uniform.

# Fix a few acronyms that title-case breaks (e.g. 'Bmw' -> 'BMW')

df['make'] = df['make'].replace({'Bmw': 'BMW', 'Gmc': 'GMC', 'Dot': 'DOT'})
# The Dictionary {'Bmw': 'BMW', ...}: This is a direct swap map. 
# The incorrect, title-cased words are on the left, and the proper all-caps acronyms are on the right.
# .replace(): Pandas looks through the make column. If it finds an exact match for "Bmw", it flips it back to "BMW". 
# If a row says "Ford", it doesn't match anything in the dictionary, so pandas completely ignores it and leaves it alone.

print(f"Unique makes after cleanup: {df['make'].nunique()}")
```

```python
# Count duplicate records
dup_count = df.duplicated().sum()
# df.duplicated(): This method scans the DataFrame row by row from top to bottom. By default, it looks at every single column. 
# The first time it sees a unique row, it marks it as False. 
# If it encounters a row further down that perfectly matches a previous row in every column, it flags it as True.
# .sum(): Since Python treats True as 1 and False as 0, summing the Boolean result counts exactly how many duplicate rows exist.

print(f"Number of duplicate records: {dup_count}")
# It is a feature that allows you to easily inject variables directly into the middle of a text string without having to break it up using plus signs (+) or commas.

# Remove duplicates if any are present
df = df.drop_duplicates()
# df.drop_duplicates(): This method scans your DataFrame row by row. 
# If it finds two or more rows that match exactly across every single column, it keeps the very first one it encounters and deletes the rest.

print(f"Shape after removing duplicates: {df.shape}")
# It is a feature that allows you to easily inject variables directly into the middle of a text string without having to break it up using plus signs (+) or commas.
```

```python
#Task 2.1 – Average, Minimum, and Maximum Selling Price

# Calculate the average, minimum, and maximum selling price
price_stats = df['sellingprice'].agg(['mean', 'min', 'max'])
# This isolates the sellingprice column from your DataFrame so that the math only runs on the actual numbers people paid for the cars.
# The .agg() function is short for aggregate. 
# It allows you to run multiple mathematical operations simultaneously on a single column instead of having to write three separate lines of code.
# By passing a list of text strings inside the brackets, you tell pandas exactly which statistics to compute:
        # 'mean': Calculates the mathematical average of all selling prices.
        # 'min': Finds the lowest transaction price in the dataset (which, according to your notebook, is $1.00).
        # 'max': Finds the absolute highest transaction price in the dataset (the ultra-luxury vehicle or supercar sitting at $230,000.00).

print(f"Average selling price: ${price_stats['mean']:,.2f}")
# .2 (The Precision): This sets the round-off limit strictly to two decimal places. This is ideal for money, ensuring you don't get messy floating-point outputs like
# f (The Float Type): This locks the data type representation as a "Fixed-point floating number" so the decimal formatting rule applies properly.

# The First f (Right before the quotation mark)
# This f stands for Format.
# It tells Python: "Hey, treat this whole line as an f-string. Look inside the curly braces {} and swap out the variable name for the actual number."

print(f"Minimum selling price: ${price_stats['min']:,.2f}")
print(f"Maximum selling price: ${price_stats['max']:,.2f}")
```

```python
#Task 2.11 – Which State Consistently Has Higher Car Prices for Newer Cars (Year > 2013)?

# Restrict to newer cars (year > 2013) and compare average price across states
newer_cars = df[df['year'] > 2013]
# What it means: We only want to look at the shiny, newer cars.
# The Action: You go through your giant toy box and throw away (or ignore) any car made in 2013 or older. You only keep the cars made in 2014, 2015, and newer.

state_avg_price = newer_cars.groupby('state')['sellingprice'].agg(['mean', 'count']).sort_values('mean', ascending=False)
# groupby('state') (The Piles): You take your pile of newer cars and sort them into mini-piles based on the state they came from. 
# All the cars from California go into a "CA" pile, all the ones from Ohio go into an "OH" pile, and so on.
# .agg(['mean', 'count']) (The Math): For each state's pile, you do two things:
# count: You count exactly how many cars are in that pile.
# mean: You add up all their prices and find the average price for that state.
# sort_values('mean', ascending=False) (The Lineup): You line up the state piles in order from the most expensive average price to the cheapest average price.

print("Top 10 states by average price for cars newer than 2013:")
# What it means: "Show me the top of the list!"
# The Action: You look at your freshly sorted lineup and print out just the top 10 states with the highest average prices.

state_avg_price.head(10)
```

```python
# Task 3.7 – Box Plot: Selling Price Distribution by Color

# Restrict to the 10 most common colors for readability
top_colors = df['color'].value_counts().head(10).index
# df['color'].value_counts(): This count how many cars there are for each color and auto sorts them from most common to least common (eg Black:500, White:480,.., Purple:2).
# .head(10): Grabs just the top 10 most frequent colors.
# .index: Instead of keeping the counts, this extracts just the names of those top 10 colors (e.g., ['black', 'white', 'silver', ... ]).

df_top_colors = df[df['color'].isin(top_colors)]
# What it does: It filters your entire dataset. It creates a new DataFrame (df_top_colors) that only includes rows where the car's color matches one of our top 10 colors.
# Any car with a rare color is left out.

# Box plot of selling price grouped by color, including outliers
plt.figure(figsize=(12,6))
sns.boxplot(data=df_top_colors, x='color', y='sellingprice', color='orange')
# figsize=(12,6): Sets up a slightly wider canvas so 10 distinct boxes have plenty of room to breathe.
# sns.boxplot(...): This creates the plot.
# x='color' splits the data into 10 columns on the horizontal axis (one for each color).
# y='sellingprice' maps the price vertically.
# color='orange' tints the boxes a uniform light blue.

plt.xlabel('Color')
plt.ylabel('Selling Price ($)')
plt.title('Selling Price Distribution by Color (Top 10 Colors) — with Outliers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# Just like the bar charts, this labels the axes, titles the graph, tilts the color names at 45 degrees so they don't overlap, and renders the plot cleanly.
```

```python
# Remove outliers per color group using the IQR method
q1 = df_top_colors.groupby('color', observed=True)['sellingprice'].transform(lambda s: s.quantile(0.25))
q3 = df_top_colors.groupby('color', observed=True)['sellingprice'].transform(lambda s: s.quantile(0.75))
# What it does: It finds the 25th percentile ($Q_1$, the bottom of the box) and the 75th percentile ($Q_3$, the top of the box) for each color.
# transform(): This is the secret weapon here. Instead of collapsing our dataset into a tiny summary table, transform() calculates the value for each color 
# and stamps it onto every single row matching that color. This lets us easily compare individual car prices against their group's limits.

iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
# iqr: The height of the box ($Q_3 - Q_1$), which represents the middle 50% of the data.
# The Formula: The standard mathematical rule for detecting outliers is to go 1.5 times the height of the box below the bottom and above the top. 
# Anything outside this range is officially deemed "too extreme" (an outlier).

df_no_outliers = df_top_colors[(df_top_colors['sellingprice'] >= lower_bound) & (df_top_colors['sellingprice'] <= upper_bound)]
# What it does: It keeps only the cars whose sellingprice sits safely between their specific color's lower_bound and upper_bound.
# The outliers are left behind, leaving us with a clean dataset called df_no_outliers.

# Re-plot the box plot without outliers
plt.figure(figsize=(12,6))
sns.boxplot(data=df_no_outliers, x='color', y='sellingprice', color='orange')
plt.xlabel('Color')
plt.ylabel('Selling Price ($)')
plt.title('Selling Price Distribution by Color (Top 10 Colors) — Outliers Removed')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# Finally, we redraw the exact same box plot using our filtered data, this time swapping the color to 'orange'.
```
